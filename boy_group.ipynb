```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import plotly.graph_objs as go
import plotly.offline as py
import plotly.express as px

import os, time, sys
import warnings
warnings.filterwarnings('ignore')
```

```python
path='/content/drive/MyDrive/Analytic/kpop/kpop_idol'
df=pd.read_csv(path+'/kpop_idols.csv',encoding='utf-8')
boy_group=pd.read_csv(path+'/kpop_idols_boy_groups.csv',encoding='utf-8')
girl_group=pd.read_csv(path+'/kpop_idols_girl_groups.csv',encoding='utf-8')
```

```python
df.isna().sum()
```

```python
fig = plt.figure(figsize=[10,7])
sns.countplot(boy_group['Members'], color=sns.xkcd_rgb['denim blue'])
plt.title('Kpop Idols Boy Groups')
plt.show()
```

```python
fig = px.pie(boy_group,
             values="Members",
             names="Orig. Memb.",
             template="seaborn")
fig.update_traces(rotation=90, pull=0.05, textinfo="percent+label")
fig.show()
```

```python
boy_group[['Short', 'Company', 'Fanclub Name']] = boy_group[['Short', 'Company', 'Fanclub Name']].fillna('NA')

fig = px.bar(boy_group,
             x='Korean Name',
             y='Members',
             orientation='h',
             color='Company',
             title='Kpop Idols Boy Groups',
             opacity=0.8,
             color_discrete_sequence=px.colors.diverging.RdBu,
             template="plotly_dark"
            )
fig.update_xaxes(range=[0,35])
fig.show()
```
